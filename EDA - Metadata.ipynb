```python
# EDA
import pandas as pd
EDA_data = pd.read_pickle('meta_dtm.pkl')
# Transpose so the words are rows, this way I can filter/sort by podcast easier 
EDA_data = EDA_data.transpose()
EDA_data.head
```

```python
# As part of EDA let's explore the most common words and see if we need to clean any out
top_dict = {}
for c in EDA_data.columns:
    top = EDA_data[c].sort_values(ascending=False).head(30)
    top_dict[c] = list(zip(top.index, top.values))
    
top_dict
```

```python
from collections import Counter

# Pull out top 30 words for each podcast
words = []
for podcast in EDA_data.columns:
    top = [word for (word, count) in top_dict[podcast]]
    for t in top:
        words.append(t)

words
```

```python
# Look at most common words in all podcast transcripts
# None of these look like stopwords, except podcast, so let's just add that to the stopwords list
Counter(words).most_common()
```

```python
# Set a limit for which of these common words to add to stop word list
# Setting at >19 podcasts included as stop words, after that we start to get to words like 'money'=12, 'story'=10, 'home'=8
add_stop_words = [word for word, count in Counter(words).most_common() if count > 80]
add_stop_words
```

```python
# Update the Doc-Term Matrix to exclude the new stop words added
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

# Read in cleaned data
data_clean = pd.read_pickle('meta_data_clean.pkl')

# Add new stop words
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

# Re-create Doc-Term Matrix with new stop word list
cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(data_clean.metacorpus)
data_stop = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_stop.index = data_clean.index

# Pickle it again in this form for use in modeling
import pickle
pickle.dump(cv, open("meta_cv_stop.pkl", "wb"))
data_stop.to_pickle("meta_dtm_stop.pkl")
```

```python
IDs = list(range(1, 301))
#Unique Words
#Identify non-zero words in the Doc-Term Metrix
unique_list = []
for podcast in EDA_data.columns:
    uniques = EDA_data[podcast].nonzero()[0].size
    unique_list.append(uniques)
    
# Create a new dataframe that contains this unique word count
data_words = pd.DataFrame(list(zip(IDs, unique_list)), columns=['podcast', 'unique_words'])
data_unique_sort = data_words.sort_values(by='unique_words')
data_unique_sort
```

```python
# Total and unique words counts to excel for summaries
# Find total # words in each podcast to add to uniques
# 
total_list = []
for podcast in EDA_data.columns:
    totals = sum(EDA_data[podcast])
    total_list.append(totals)
data_words['total_words'] = total_list  
```

```python
# Export to CSV
data_words.to_csv('meta_totalwords.csv')
```
