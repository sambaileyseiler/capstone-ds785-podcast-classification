```python
import re
import string

def clean_text_round1(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words...'''
    text = text.lower() # lowersace everything
    text = re.sub('\[.*?\]', '', text) # remove square brackets timestamps
    text = re.sub('[%s]' % re.escape(string.punctuation),'', text) #remove punctuation marks
    text = re.sub('\w*\d\w*', '', text) # remove words that contain numbers
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
# EDA
import pandas as pd
EDA_data = pd.read_pickle('dtm.pkl')
# Transpose so the words are rows, this way I can filter/sort by podcast easier 
EDA_data = EDA_data.transpose()
EDA_data.head
```

```python
# As part of EDA let's explore the most common words and see if we need to clean any out
top_dict = {}
for c in EDA_data.columns:
    top = EDA_data[c].sort_values(ascending=False).head(30)
    top_dict[c] = list(zip(top.index, top.values))
    
top_dict
```

```python
from collections import Counter

# Pull out top 30 words for each podcast
words = []
for podcast in EDA_data.columns:
    top = [word for (word, count) in top_dict[podcast]]
    for t in top:
        words.append(t)

words
```

```python
# Look at most common words in all podcast transcripts
# Some of these seem like they could be considered stop words, common throughout all podcasts, so let's exclude them
Counter(words).most_common()
```

```python
# Set a limit for which of these common words to add to stop word list
# Setting at >19 podcasts included as stop words, after that we start to get to words like 'money'=12, 'story'=10, 'home'=8
add_stop_words = [word for word, count in Counter(words).most_common() if count > 19]
add_stop_words
```

```python
# Update the Doc-Term Matrix to exclude the new stop words added
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

# Read in cleaned data
data_clean = pd.read_pickle('data_clean.pkl')

# Add new stop words
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

# Re-create Doc-Term Matrix
cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(data_clean.transcript)
data_stop = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_stop.index = data_clean.index

# Pickle it again in this form
import pickle
pickle.dump(cv, open("cv_stop.pkl", "wb"))
data_stop.to_pickle("dtm_stop.pkl")
```
